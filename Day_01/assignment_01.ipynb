```python
sent = 'They told that their ages are 23, 25 and 29 respectively.'
```

```python
# 1. Find the average of ages.
ages = [int(word) for word in sent.split() if word.isdigit()]
average_age = sum(ages) / len(ages) if ages else 0
print(average_age)
```

```python
# 2. Count total number of words starts with 't' / 'T'
words = sent.split()
count = 0

for word in words:
    if word.lower().startswith('t'):
        count += 1

print("Words starting with T/t:", count)
print(words)
```

```python
# 3. Count total punctuation symbols.
import string
count_punct = 0
for ch in sent:
    if ch in string.punctuation:
        count_punct+=1
print('total punctuation symbols:',count_punct)
```

```python
# 4. Print the words in capital letter which ends with 'd' 
for word in words:
    if word.endswith('d'):
        print(word.upper())
```

```python
# 5. Merge all the words and print (no punctuations)
result = ""

for ch in sent:
    if ch not in string.punctuation:
        result += ch

result = result.replace(" ", "")

print(result)
```
